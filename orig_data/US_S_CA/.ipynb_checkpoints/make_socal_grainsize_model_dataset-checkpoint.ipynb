```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
from tqdm import tqdm
```

```python
# ! pip install geopy
from geopy.geocoders import Nominatim
```

```python
mpa = pd.read_csv('MPA_Data_Fin.csv')
```

```python
mpa.keys()
```

```python
mpa = mpa[['PROJECT', 'SITE', 'LEVEL', 'd50']]
mpa = mpa.dropna()
```

```python
mpa.keys()
```

```python
mpa
```

```python
mpa['SITE'][0]
```

```python
locator = Nominatim(user_agent="myGeocoder")
location = locator.geocode(mpa['SITE'][0])
print(location.latitude)
print(location.longitude)
```

```python
LATS = []
LONS = []
for s in tqdm(mpa['SITE']):
    location = locator.geocode(s)
    try:
        LONS.append(location.longitude)
        LATS.append(location.latitude)
    except:
        LONS.append(np.nan)
        LATS.append(np.nan)
```

```python
mpa['Lat'] = LATS
mpa['Lon'] = LONS
mpa
```

```python
mpa = mpa.dropna()
mpa
```

```python
tmp = mpa['LEVEL'].values.copy()
tmp[tmp=='HTS'] = 'MB'
tmp[tmp=='WTO'] = 'BF'
```

```python
mpa['LEVEL'] = tmp
mpa
```

```python
mpa.rename(columns = {'PROJECT':'dataset', 'LEVEL':'Ele', 'd50': 'Grain_Size'}, inplace = True)
mpa
```

```python
mpa = mpa[['Lat','Lon','dataset', 'Ele', 'Grain_Size']]
mpa = mpa.dropna()
mpa
```

```python
mpa = mpa[mpa['Lat']<35]
mpa
```
