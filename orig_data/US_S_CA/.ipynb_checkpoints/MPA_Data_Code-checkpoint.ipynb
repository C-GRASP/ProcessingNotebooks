```python
import ipyleaflet
import  os
import pandas as pd
import geopandas as gpd
import numpy as np
import geocoder

import scipy
```

```python
#read raw data

mpa=pd.read_csv('/home/williamcoast/Desktop/SandData/mpa_data.csv')
```

```python
#interpolate from phi measurements

i=0
for i in range(0,len(mpa)):


    prcs = [5,10,16,25,30,50,75,84,90,95]

    grain_size_bins=[ float(mpa.columns[10]),
                    float(mpa.columns[11]),
                     float(mpa.columns[12]),
                     float(mpa.columns[13]),
                     float(mpa.columns[14]),
                     float(mpa.columns[15]),
                     float(mpa.columns[16]),
                     float(mpa.columns[17]),
                     float(mpa.columns[18]),
                     float(mpa.columns[19]),
                     float(mpa.columns[20]),
                     float(mpa.columns[21]),
                     float(mpa.columns[22]),
                     float(mpa.columns[23]),
                     float(mpa.columns[24])]
    grain_size_frequencies=[mpa.loc[i,mpa.columns[10]],
                            mpa.loc[i,mpa.columns[11]],
                            mpa.loc[i,mpa.columns[12]],
                            mpa.loc[i,mpa.columns[13]],
                            mpa.loc[i,mpa.columns[14]],
                            mpa.loc[i,mpa.columns[15]],
                            mpa.loc[i,mpa.columns[16]],
                            mpa.loc[i,mpa.columns[17]],
                            mpa.loc[i,mpa.columns[18]],
                            mpa.loc[i,mpa.columns[19]],
                            mpa.loc[i,mpa.columns[20]],
                            mpa.loc[i,mpa.columns[21]],
                            mpa.loc[i,mpa.columns[22]],
                            mpa.loc[i,mpa.columns[23]],
                            mpa.loc[i,mpa.columns[24]]]
    
    
    
    prc_values = np.interp(prcs,np.hstack((0,np.cumsum(grain_size_frequencies))), np.hstack((0,grain_size_bins)))


    mpa.loc[i,["d5"]] = prc_values[0]
    mpa.loc[i,["d10"]] = prc_values[1]
    mpa.loc[i,["d16"]] = prc_values[2]
    mpa.loc[i,["d25"]] = prc_values[3]
    mpa.loc[i,["d30"]] = prc_values[4]
    mpa.loc[i,["d50"]] = prc_values[5]
    mpa.loc[i,["d75"]] = prc_values[6]
    mpa.loc[i,["d84"]]= prc_values[7]
    mpa.loc[i,["d90"]] = prc_values[8]
    mpa.loc[i,["d95"]]= prc_values[9]
    
    
    
    d5=mpa.loc[i,["d5"]].astype(float).item()
    d10=mpa.loc[i,["d10"]].astype(float).item() 
    d16=mpa.loc[i,["d16"]].astype(float).item()  
    d25=mpa.loc[i,["d25"]].astype(float).item() 
    d30=mpa.loc[i,["d30"]].astype(float).item()
    d50=mpa.loc[i,["d50"]].astype(float).item()
    d75=mpa.loc[i,["d75"]].astype(float).item()
    d84=mpa.loc[i,["d84"]].astype(float).item()
    d90=mpa.loc[i,["d90"]].astype(float).item()
    d95=mpa.loc[i,["d95"]].astype(float).item()
    
    


    i=i+1
    
    



```

```python
#convert phi measurements to mm
mpa['d5']=10**((-1*mpa['d5'])/3.322)    
mpa['d10']=10**((-1*mpa['d10'])/3.322)    
mpa['d16']=10**((-1*mpa['d16'])/3.322)    
mpa['d25']=10**((-1*mpa['d25'])/3.322)    
mpa['d50']=10**((-1*mpa['d50'])/3.322)    
mpa['d75']=10**((-1*mpa['d75'])/3.322)    
mpa['d84']=10**((-1*mpa['d84'])/3.322)    
mpa['d90']=10**((-1*mpa['d90'])/3.322)    
mpa['d95']=10**((-1*mpa['d95'])/3.322) 
```

```python
#add stats
mpa['Mean']=np.exp(((np.log(mpa['d16']))+(np.log(mpa['d50']))+(np.log(mpa['d84'])))/3)
mpa['Std']=np.exp(((((np.log(mpa['d16']))-(np.log(mpa['d84'])))/4)+(((np.log(mpa['d5']))-(np.log(mpa['d95'])))/4)))
mpa['Skewness']=np.exp(((((np.log(mpa['d16']))+(np.log(mpa['d84']))-2*(np.log(mpa['d50'])))/(2*((np.log(mpa['d84']))-(np.log(mpa['d16'])))))
                         +(((np.log(mpa['d5']))+(np.log(mpa['d95']))-2*(np.log(mpa['d50'])))/(2*((np.log(mpa['d25']))-(np.log(mpa['d5'])))))))
mpa['Kurtosis']=(((np.log(mpa['d5']))-(np.log(mpa['d95'])))/(2.44*((np.log(mpa['d25']))-(np.log(mpa['d95'])))))
```

```python
#project and contact info
mpa['contact']='Karina Nielsen'
mpa['Dataset']='2011-2013 MPA Study'
```

```python
#geocode lat/lons
mpa['SITE']=mpa['SITE']+', California' 
for i in range(0,len(mpa)):
    try:
        mpa['latitude'].iloc[i]=geocoder.osm(mpa['SITE'].iloc[i]).latlng[0]
        mpa['longitude'].iloc[i]=geocoder.osm(mpa['SITE'].iloc[i]).latlng[1]
    except:
        pass
```
