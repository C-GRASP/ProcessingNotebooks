```python
waves0 = gpd.read_file('wave_data/CDIPS_CA_yearly_region_0.geojson', driver='GEOJSON')#.mean(axis=0)
waves1 = gpd.read_file('wave_data/CDIPS_CA_yearly_region_1.geojson', driver='GEOJSON')#.mean(axis=0)
waves2 = gpd.read_file('wave_data/CDIPS_CA_yearly_region_2.geojson', driver='GEOJSON')#.mean(axis=0)
waves3 = gpd.read_file('wave_data/CDIPS_CA_yearly_region_3.geojson', driver='GEOJSON')#.mean(axis=0)
waves4 = gpd.read_file('wave_data/CDIPS_CA_yearly_region_4.geojson', driver='GEOJSON')#.mean(axis=0)
waves5 = gpd.read_file('wave_data/CDIPS_CA_yearly_region_5.geojson', driver='GEOJSON')#.mean(axis=0)
waves6 = gpd.read_file('wave_data/CDIPS_CA_yearly_region_6.geojson', driver='GEOJSON')#.mean(axis=0)
```

```python
waves0
```

```python
waves1.mean(axis=0).T
```

```python
tmp0=pd.DataFrame(waves0.mean(axis=0), columns=['0'])
```

```python
tmp1=pd.DataFrame(waves1.mean(axis=0), columns=['1'])
```

```python
tmp2=pd.DataFrame(waves2.mean(axis=0), columns=['2'])
tmp3=pd.DataFrame(waves3.mean(axis=0), columns=['3'])
tmp4=pd.DataFrame(waves4.mean(axis=0), columns=['4'])
tmp5=pd.DataFrame(waves5.mean(axis=0), columns=['5'])
tmp6=pd.DataFrame(waves6.mean(axis=0), columns=['6'])
```

```python
tmp0
```

```python
tmp1
```

```python
waves = tmp0.join(tmp1)
waves = waves.join(tmp2)
waves = waves.join(tmp3)
waves = waves.join(tmp4)
waves = waves.join(tmp5)
waves = waves.join(tmp6)
```

```python
waves
```

```python
waves = waves.T.rename(columns={"metaLatitude": "Lat", "metaLongitude": "Lon"})
```

```python
waves_gdf = gpd.GeoDataFrame(waves, geometry=gpd.points_from_xy(waves.Lon, waves.Lat))
```

```python
waves_gdf.plot()
gdf.plot()
```

```python
waves_gdf[['Hs', 'Tp', 'Ta', 'Dp', 'Da', 'Ea', 'Es',
       'QC', 'metaWaterDepth', 'geometry']]
```

```python
gdf.keys()
```

```python
gdf_gs_waves = gpd.sjoin_nearest(gdf[['Dep', 'Ele','dataset','Grain_Size', 'geometry']], waves_gdf[['Hs', 'Tp', 'Ta', 'Dp', 'Da', 'Ea', 'Es',
       'metaWaterDepth', 'geometry']], how='left')
```

```python
df_gs_waves = gdf_gs_waves[['Dep', 'Ele','dataset','Grain_Size','Hs', 'Tp', 'Ta', 'Dp', 'Da', 'Ea', 'Es']]

df_gs_waves.to_csv('SoCal_GrainSize_waves_merged.csv')
```
