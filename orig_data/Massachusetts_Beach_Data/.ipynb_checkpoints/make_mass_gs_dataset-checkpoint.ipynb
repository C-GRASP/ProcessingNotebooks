```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
from tqdm import tqdm
```

```python
dat = pd.read_csv('UMASS_cgrasp.csv')
```

```python
dat.keys()
```

```python
dat = dat[['latitude', 'longitude', 'd50','Location_Type']]
dat = dat.dropna()
```

```python
locs = np.unique(dat['Location_Type'])
locs
```

```python
DF = []
for k in np.unique(dat['Location_Type']):
    DF.append(dat[dat['Location_Type']==k].copy())
```

```python
lens = [len(k) for k in DF]
lens
```

```python
for c,l in enumerate(lens):
    print(locs[c]+' :')
    print(l)
```

```python
dfHT = DF[6]
dfLT = DF[7]
dfMT = DF[8]
```

```python
dfHT
```

```python
dfHT.to_csv('UMASS_hightide.csv')
dfLT.to_csv('UMASS_lowtide.csv')
dfMT.to_csv('UMASS_midtide.csv')
```

```python
gdfHT = gpd.GeoDataFrame(
    dfHT, geometry=gpd.points_from_xy(dfHT.longitude, dfHT.latitude))
```

```python
gdfLT = gpd.GeoDataFrame(
    dfLT, geometry=gpd.points_from_xy(dfLT.longitude, dfLT.latitude))

gdfMT = gpd.GeoDataFrame(
    dfMT, geometry=gpd.points_from_xy(dfMT.longitude, dfMT.latitude))
```

```python
plt.subplot(221)
plt.hist(gdfHT['d50'], bins=np.linspace(0,2,10))
plt.subplot(222)
plt.hist(gdfMT['d50'], bins=np.linspace(0,2,10))
plt.subplot(223)
plt.hist(gdfLT['d50'], bins=np.linspace(0,2,10))
```

```python
### make waves
```

```python
waves = pd.read_csv('Massachusetts_Beaches_WaveHeight_Tides.csv')
waves.keys()
```

```python
waves = waves[['Latitude', 'Longitude','Hsig: prior 30 days 12 hr-avg max ', 'Mean Tidal Range']].copy()

waves = waves.drop_duplicates()
waves = waves.dropna()
len(waves)
waves = waves.drop(0)

waves.rename(columns = {'Latitude':'latitude', 'Longitude':'longitude', 'Hsig: prior 30 days 12 hr-avg max ': 'hs_mean', 'Mean Tidal Range': 'mstr'}, inplace = True)
```

```python
waves
```

```python
waves_gdf = gpd.GeoDataFrame(waves, geometry=gpd.points_from_xy(waves.longitude, waves.latitude))
waves_gdf
```

```python
gdfHT_waves = gpd.sjoin_nearest(gdfHT,waves_gdf, how='left',  distance_col='distance')
gdfHT_waves
```

```python
plt.hist(gdfHT_waves['d50'], bins=np.linspace(0,2,10))
```

```python
gdfLT_waves = gpd.sjoin_nearest(gdfLT,waves_gdf, how='left',  distance_col='distance')
gdfLT_waves
```

```python
plt.hist(gdfLT_waves['d50'], bins=np.linspace(0,2,10))
```

```python
gdfMT_waves = gpd.sjoin_nearest(gdfMT,waves_gdf, how='left',  distance_col='distance')
gdfMT_waves
```

```python
plt.hist(gdfMT_waves['d50'], bins=np.linspace(0,2,10))
```

```python
len(gdfMT)
```

```python
### make slope
```

```python
slopes = pd.read_csv('Massachusetts_Beaches_GrainSize_Slope.csv')
slopes.keys()
```

```python
slopes = slopes[['Lat', 'Lon','Beach Face']].copy()

slopes = slopes.drop_duplicates()
slopes = slopes.dropna()
len(slopes)

slopes = slopes.drop(0)

slopes.rename(columns = {'Lat':'latitude', 'Lon':'longitude','Beach Face': 'beach_slope_average'}, inplace = True)
```

```python
slopes_gdf = gpd.GeoDataFrame(slopes, geometry=gpd.points_from_xy(slopes.longitude, slopes.latitude))
slopes_gdf
```

```python
## high tide
```

```python
gdfHT_waves = gdfHT_waves[['latitude_left', 'longitude_left', 'hs_mean', 'mstr',
       'd50', 'geometry']]
gdfHT_waves.rename(columns = {'latitude_left':'latitude', 'longitude_left':'longitude'}, inplace = True)
gdfHT_waves
```

```python
# plt.subplot(221)
plt.hist(gdfHT_waves['d50'], bins=np.linspace(0,2,10))
# plt.subplot(222)
# plt.hist(gdfMT['d50'], bins=np.linspace(0,2,10))
# plt.subplot(223)
# plt.hist(gdfLT['d50'], bins=np.linspace(0,2,10))
```

```python
gdfHT_waves_slopes = gpd.sjoin_nearest(gdfHT_waves,slopes_gdf, how='left',  distance_col='distance')
gdfHT_waves_slopes
```

```python
gdfHT_waves_slopes = gdfHT_waves_slopes[['latitude_left', 'longitude_left', 'hs_mean', 'mstr',
       'd50', 'beach_slope_average']]
gdfHT_waves_slopes.rename(columns = {'latitude_left':'latitude', 'longitude_left':'longitude'}, inplace = True)
gdfHT_waves_slopes = gdfHT_waves_slopes.drop_duplicates()

gdfHT_waves_slopes
```

```python
## mid tide
```

```python
gdfMT_waves = gdfMT_waves[['latitude_left', 'longitude_left', 'hs_mean', 'mstr',
       'd50', 'geometry']]
gdfMT_waves.rename(columns = {'latitude_left':'latitude', 'longitude_left':'longitude'}, inplace = True)
gdfMT_waves
```

```python
gdfMT_waves_slopes = gpd.sjoin_nearest(gdfMT_waves,slopes_gdf, how='left',  distance_col='distance')
gdfMT_waves_slopes
```

```python
gdfMT_waves_slopes = gdfMT_waves_slopes[['latitude_left', 'longitude_left', 'hs_mean', 'mstr',
       'd50', 'beach_slope_average']]
gdfMT_waves_slopes.rename(columns = {'latitude_left':'latitude', 'longitude_left':'longitude'}, inplace = True)
gdfMT_waves_slopes = gdfMT_waves_slopes.drop_duplicates()

gdfMT_waves_slopes
```

```python
## low tide
```

```python
gdfLT_waves = gdfLT_waves[['latitude_left', 'longitude_left', 'hs_mean', 'mstr',
       'd50', 'geometry']]
gdfLT_waves.rename(columns = {'latitude_left':'latitude', 'longitude_left':'longitude'}, inplace = True)
gdfLT_waves
```

```python
gdfLT_waves_slopes = gpd.sjoin_nearest(gdfLT_waves, slopes_gdf, how='left',  distance_col='distance')
gdfLT_waves_slopes
```

```python
gdfLT_waves_slopes = gdfLT_waves_slopes[['latitude_left', 'longitude_left', 'hs_mean', 'mstr',
       'd50', 'beach_slope_average']]
gdfLT_waves_slopes.rename(columns = {'latitude_left':'latitude', 'longitude_left':'longitude'}, inplace = True)
gdfLT_waves_slopes = gdfLT_waves_slopes.drop_duplicates()

gdfLT_waves_slopes
```

```python
plt.subplot(221)
plt.hist(gdfHT_waves_slopes['d50'], bins=np.linspace(0,2,10))
plt.subplot(222)
plt.hist(gdfMT_waves_slopes['d50'], bins=np.linspace(0,2,10))
plt.subplot(223)
plt.hist(gdfLT_waves_slopes['d50'], bins=np.linspace(0,2,10))
```
