```python
slopes =pd.DataFrame(gp.read_file('data/beach_slopes/US_East_coast_slopes_by_transect.geojson').to_crs(epsg=4326))
```

```python
slopes['points'] = slopes.apply(lambda x: [y for y in x['geometry'].coords], axis=1)
```

```python
latlon = np.vstack([np.mean(s,axis=0) for s in slopes.points.values])

cs_lons = latlon[:,0]
cs_lats = latlon[:,1]
```

```python
slopes['latitude'] = cs_lats
slopes['longitude'] = cs_lons
```

```python
gs_slope_waves = gs_slope_waves.dropna()
len(gs_slope_waves)
```

```python
plt.figure(figsize=(8,8))
plt.loglog(bujan_dat['Clast size (mm)'], bujan_dat['Slope (tan β)'],'ko', label='Bujan et al. (2019)')
plt.loglog(gs_slope_waves['d50'], gs_slope_waves['tanBeta'],'rd', label='Atlantic C-GRASP')
plt.xlabel('Clast size (mm)')
plt.ylabel('Slope (tan β)')

a = -0.154
b = -0.145
c = 0.268
ln = a*( bujan_dat['Clast size (mm)'] - 0.125)**b + c

plt.plot(np.sort(bujan_dat['Clast size (mm)']), np.sort(ln),'m', label=r'$-.154(D50 - .125)^{-.145} + .268$')

plt.legend()

plt.savefig('CGRASP_d50-slope-allsamples.jpg', dpi=300)
```

```python
plt.figure(figsize=(8,8))
plt.loglog(bujan_dat['Clast size (mm)'], bujan_dat['Slope (tan β)'],'ko', markersize=1, label='Bujan et al. (2019)')
plt.scatter(gs_slope_waves['d50'], gs_slope_waves['tanBeta'],20,gs_slope_waves['hs_mean'],label='Atlantic C-GRASP')
plt.xlabel('Clast size (mm)')
plt.ylabel('Slope (tan β)')
cb = plt.colorbar()

cb.set_label(r'Mean $H_s$ (m)')

a = -0.154
b = -0.145
c = 0.268
ln = a*( bujan_dat['Clast size (mm)'] - 0.125)**b + c

plt.plot(np.sort(bujan_dat['Clast size (mm)']), np.sort(ln),'m', label=r'$-.154(D50 - .125)^{-.145} + .268$')

plt.xlim(0.1,2)

plt.legend()

plt.savefig('CGRASP_d50-slope-allsamples-waveHs.jpg', dpi=300)
```

```python
plt.figure(figsize=(8,8))
plt.loglog(bujan_dat['Clast size (mm)'], bujan_dat['Slope (tan β)'],'ko', markersize=1, label='Bujan et al. (2019)')
plt.scatter(gs_slope_waves['d50'], gs_slope_waves['tanBeta'],20,gs_slope_waves['tp'],label='Atlantic C-GRASP')
plt.xlabel('Clast size (mm)')
plt.ylabel('Slope (tan β)')
cb = plt.colorbar()

cb.set_label(r'Mean $T_p$ (s)')

a = -0.154
b = -0.145
c = 0.268
ln = a*( bujan_dat['Clast size (mm)'] - 0.125)**b + c

plt.plot(np.sort(bujan_dat['Clast size (mm)']), np.sort(ln),'m', label=r'$-.154(D50 - .125)^{-.145} + .268$')

plt.xlim(0.1,2)

plt.legend()

plt.savefig('CGRASP_d50-slope-allsamples-waveTp.jpg', dpi=300)
```

```python
gs_w_slope_data = gdf_sand.sjoin_nearest(gdf_slopes, distance_col="distances", how="right")
```

```python
## within 100m
```

```python
merged_filtered = gs_w_slope_data[gs_w_slope_data['distances']<.0015] 
```

```python
merged_lats = merged_filtered['latitude_left']
merged_lons = merged_filtered['longitude_left']
```
