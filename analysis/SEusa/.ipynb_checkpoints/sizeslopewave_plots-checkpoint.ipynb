```python
g=9.8
h=10

#compute wavelength
# period and wavelength are related as λ = CT
# For shallow water waves, the phase speed, C = (gh)^0.5
lam = df['hs_mean']/(df['tp']*((g*h)**0.5))
```

### Gradient boosting regression model

```python
#https://scikit-learn.org/stable/auto_examples/ensemble/plot_gradient_boosting_regression.html
```

```python

params = {
    "n_estimators": 500,
    "max_depth": 4,
    "min_samples_split": 10, #5,
    "learning_rate": 0.01,
    "loss": "squared_error",
}
```

#### waves only model

```python
feature_names1 = ['Dir','Hs','Tp'] #'Hmax''Lamda',
```

```python
X = np.stack((df['dir'],df['hs_mean'],df['tp']))
X.shape

X_train1, X_test1, y_train1, y_test1 = train_test_split(
    X.T, df['d50'], test_size=0.5, random_state=2022
)
```

```python
model1 = ensemble.GradientBoostingRegressor(**params)
model1.fit(X_train1, y_train1)

mse1 = mean_squared_error(y_test1, model1.predict(X_test1))
print("The mean squared error (MSE) on test set: {:.4f}".format(mse1))
```

#### morpho  model

```python
feature_names3 = ['Slope', 'Trend'] #'Hmax','Lamda'
```

```python
X = np.stack((df['beach_slope'],df['trend']))
#lam,df['hs_max'],
X.shape

X_train3, X_test3, y_train3, y_test3 = train_test_split(
    X.T, df['d50'], test_size=0.5, random_state=2022
)
```

```python
model3 = ensemble.GradientBoostingRegressor(**params)
model3.fit(X_train3, y_train3)

mse3 = mean_squared_error(y_test3, model3.predict(X_test3))
print("The mean squared error (MSE) on test set: {:.4f}".format(mse3))
```
