```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.model_selection import train_test_split
```

```python
def prc_err(y_obs,y_est):
    return np.mean(100*(np.abs(y_obs-y_est)/y_obs))
```

```python
def Rsquared(y_obs, y_est):
    r2 = np.min(np.corrcoef(y_obs, y_est))**2
    if r2>1:
        r2=0
    if r2<0:
        r2=0
    return r2
```

```python
def return_best_model(model):
    return model['model'].best_estimator_
```

```python
def get_splits(X, df, test_size=0.6, rand_state=2022):

    X_train1, X_test1, y_train1, y_test1 = train_test_split(
        X.T, df['d50'], test_size=test_size, random_state=rand_state)
    
    return X_train1, X_test1, y_train1, y_test1 
```

```python
def mcfall(X,hs):
    if hs<=1:
        A = 3.1; n = -1.1 
    elif hs>=3:
        A = 3.9; n = -1.85         
    else:
        A = 2.1; n = -1.8             
    return (X/A)**(1/n)

def bujan(B,a,b,c):
    t1=(B-c)
    return (1/8)*(8*((t1/a))**(1/b)+1)


a = -0.154
b = -0.145
c = 0.268
```
