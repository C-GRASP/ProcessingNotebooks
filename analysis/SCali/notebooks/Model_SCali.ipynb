```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dfSB = pd.read_csv('../../../orig_data/US_SantaBarbara/SantaBarbara_grainsize.csv')
```

```python
dfSB.keys()
```

```python
plt.hist(dfSB['eMean'])
```

```python
dfSB['Ele']
```

```python
BF = dfSB[dfSB['Ele']=='BF']
O = dfSB[dfSB['Ele']=='O']
MB = dfSB[dfSB['Ele']=='MB']
BB = dfSB[dfSB['Ele']=='BB']
```

```python
len(BF)+len(MB)+len(BB)
```

```python
plt.figure(figsize=(10,10))

plt.subplot(221)
plt.hist(BF['eMean'], label='Eyeball')
plt.hist(BF['gMean'], label='Grab')
plt.legend()
plt.title('a) Beachface',loc='left')

plt.subplot(222)
plt.hist(BB['eMean'])
plt.hist(BB['gMean'])
plt.title('b) Back-beach',loc='left')

plt.subplot(223)
plt.hist(MB['eMean'])
plt.hist(MB['gMean'])
plt.title('c) Mid-beach',loc='left')

plt.subplot(224)
plt.hist(O['eMean'])
plt.hist(O['gMean'])
plt.title('d) Offshore',loc='left')

plt.savefig('../data_plots/SB_samples_by_elev.jpg', dpi=300, bbox_inches='tight')
```

```python
plt.figure(figsize=(10,10))

plt.subplot(221)
plt.hist(BF['eMean'][BF['Sea']=='S'], label='Summer')
plt.hist(BF['eMean'][BF['Sea']=='W'], label='Winter', alpha=0.5)
plt.legend()
plt.title('a) Beachface',loc='left')

plt.subplot(222)
plt.hist(BB['eMean'][BB['Sea']=='S'], alpha=0.5)
plt.hist(BB['eMean'][BB['Sea']=='W'], alpha=0.5)
plt.title('b) Back-beach',loc='left')

plt.subplot(223)
plt.hist(MB['eMean'][MB['Sea']=='S'], alpha=0.5)
plt.hist(MB['eMean'][MB['Sea']=='W'], alpha=0.5)
plt.title('c) Mid-beach',loc='left')

plt.subplot(224)
plt.hist(O['eMean'][O['Sea']=='S'], alpha=0.5)
plt.hist(O['gMean'][O['Sea']=='W'], alpha=0.5)
plt.title('d) Offshore',loc='left')

plt.savefig('../data_plots/SB_samples_by_season.jpg', dpi=300, bbox_inches='tight')
```
